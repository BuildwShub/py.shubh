```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

quotes_info = []

for x in range(1,11):

  website = (f"https://quotes.toscrape.com/page/{x}/")

  r = requests.get(website)

  soup = BeautifulSoup(r.content, "html.parser")
  quotes = soup.find_all("div", class_="quote")

  for item in quotes:
      quote_text = item.find("span", class_="text").text.strip()
      author_name = item.find("small", class_="author").text.strip()
      quotes_link = "https://quotes.toscrape.com" + item.find("a")["href"]
      tags = item.find_all("a", class_="tag")
      tag_list = [tag.text for tag in tags]


      all_quotes = {
          "Quote": quote_text,
          "Name": author_name,
          "Link": quotes_link,
          "Tag" : ", ".join(tag_list)
      }

      quotes_info.append(all_quotes)

df = pd.DataFrame(quotes_info)

# Convert links into clickable hyperlinks in Excel
df['Link'] = '=HYPERLINK("' + df['Link'] + '","' + df['Link'] + '")'

df.to_excel("quotes_Info2.xlsx", index=False)

print("Excel file created successfully")
```
