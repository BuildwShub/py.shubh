```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

books_info = []

for x in range(1,51):

  website = (f"https://books.toscrape.com/catalogue/page-{x}.html")
  r = requests.get(website)

  soup = BeautifulSoup(r.content, "html.parser")
  book = soup.find_all("li", class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")

  for item in book:
      name_tag = item.find("h3").find("a")
      book_name = name_tag.text.strip()
      book_link = "https://books.toscrape.com/catalogue/" + name_tag["href"]
      price = item.find("p", class_="price_color").text.strip()

      all_products = {
          "Name": book_name,
          "Price": price,
          "Link": book_link
      }

      books_info.append(all_products)

df = pd.DataFrame(books_info)

# Convert links into clickable hyperlinks in Excel
df['Link'] = '=HYPERLINK("' + df['Link'] + '","' + df['Link'] + '")'

df.to_excel("Book_Info4.xlsx", index=False)

print("Excel file created successfully")
```
